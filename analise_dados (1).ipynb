ANALISE DE DADOS - DATASET


```python
caminho_dados = ('/content/drive/MyDrive/AD/olist_customers_dataset.csv')
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
df_dados = pd.read_csv(caminho_dados)
```

```python
top_5_estados = df_dados['customer_state'].value_counts().head(5)
print("Os 5 estados com maior volume de compras são:")
print(top_5_estados)
```

```python
import matplotlib.pyplot as plt
top_5_estados.plot(kind='bar', color='skyblue')
plt.title('Top 5 Estados por Volume de Compras')
plt.xlabel('Estado')
plt.ylabel('Quantidade de Clientes')
plt.show()
```

```python
top_10_cidades = df_dados['customer_city'].value_counts().head(10)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6)) # Aumentamos um pouco o gráfico para caber os nomes das cidades
top_10_cidades.plot(kind='bar', color='coral')
plt.title('Top 10 Cidades por Volume de Compras')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de Clientes')
plt.xticks(rotation=45) # Inclina os nomes para não ficarem amontoados
plt.show()
```
